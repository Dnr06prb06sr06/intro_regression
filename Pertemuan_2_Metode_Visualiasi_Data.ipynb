```python
data['movie_title'] = data['movie_title'].str.replace('?','')
data
```

```python
data['title_year.1'] = data['title_year'].copy()
data
```

```python
data['movie_title'] = data['movie_title'].str.replace('ÿ	','')
data
```

```python
duration = data['duration'].mean()
data['duration'].fillna (duration, inplace=True)
data
```

```python
data['DIRECTOR_facebook_likes'] = data ['DIRECTOR_facebook_likes'].str.replace('\"','')
data
```

```python
data['DIRECTOR_facebook_likes'] = pd.to_numeric(data['DIRECTOR_facebook_likes'], errors='coerce')
DIRECTOR_facebook_likes = data['DIRECTOR_facebook_likes'].mean()
data['DIRECTOR_facebook_likes'].fillna (DIRECTOR_facebook_likes, inplace=True)
data
```

```python
data['facenumber_in_poster'] = pd.to_numeric(data['facenumber_in_poster'], errors='coerce')
facenumber_in_poster = data['facenumber_in_poster'].mean()
data['facenumber_in_poster'].fillna (facenumber_in_poster, inplace=True)
data
```

```python
data['Cast_Total_facebook_likes'] = pd.to_numeric(data['Cast_Total_facebook_likes'], errors='coerce')
Cast_Total_facebook_likes = data['Cast_Total_facebook_likes'].mean()
data['Cast_Total_facebook_likes'].fillna (Cast_Total_facebook_likes, inplace=True)
data
```

```python
data['ACTOR_2_facebook_likes'] = pd.to_numeric(data['ACTOR_2_facebook_likes'], errors='coerce')
ACTOR_2_facebook_likes = data['ACTOR_2_facebook_likes'].mean()
data['ACTOR_2_facebook_likes'].fillna (ACTOR_2_facebook_likes, inplace=True)
data
```

```python
data['num_voted_users'] = pd.to_numeric(data['num_voted_users'], errors='coerce')
num_voted_users = data['num_voted_users'].mean()
data['num_voted_users'].fillna (num_voted_users, inplace=True)
data
```

```python
data['ACTOR_2_facebook_likes'] = pd.to_numeric(data['ACTOR_2_facebook_likes'], errors='coerce')
ACTOR_2_facebook_likes = data['ACTOR_2_facebook_likes'].mean()
data['ACTOR_2_facebook_likes'].fillna (ACTOR_2_facebook_likes, inplace=True)
data
```

```python
ACTOR_2_facebook_likes = data['ACTOR_2_facebook_likes'].mean()
data['ACTOR_2_facebook_likes'].fillna(ACTOR_2_facebook_likes, inplace=True)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
import numpy as np # Added this line to import the numpy module

numeric_data = data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
X = data.iloc[:,8]
X.head(10)
```

```python
y = data.iloc[:,13]
y.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=51)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
linear = LinearRegression()
linear.fit(np.array(X_train).reshape(-2,1),y_train)
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
plt.scatter(X,y)
plt.title('cast_Total_facebook_likes vs ACTOR_2_facebook_likes')
plt.xlabel('ACTOR_2_facebook_likes')
plt.ylabel('cast_Total_facebook_likes')
```

```python
# Make sure to execute this cell first to define Y_pred
Y_pred = linear.predict(np.array(X_test).reshape(-1,1))

plt.scatter(X_test,y_test)
plt.plot(X_test,Y_pred,color='black') # Now Y_pred is defined and can be used
plt.title('Cast_Total_facebook_likes vs ACTOR_2_facebook_likes')
plt.xlabel('ACTOR_2_facebook_likes')
plt.ylabel('Cast_Total_facebook_likes')
```
